### Exercício 2.4



Crie uma função que tenha como entradas:
- Inteiro **m** que corresponde às linhas de uma matriz;
- Inteiro **n** que corresponde às colunas de uma matriz;

e que retorne:
- Uma matriz **X** de shape (m, n), em que cada elemento X[i,j] seja a distância Euclidiana entre a posição [i,j] e a posição [0,0].

Dica: criar novas matrizes com as funções [*np.meshgrid*](https://github.com/lrittner/Computacao-Visual/blob/master/tutoriais/02_tutorial_numpy_1_6.ipynb) e [*np.arrange*](https://github.com/lrittner/Computacao-Visual/blob/master/tutoriais/02_tutorial_numpy_1_1.ipynb) para realizar as operações e obter os valores que comporão a matriz X.

Teste a função com os valores **m=5** e **n=4**.

```python
def euclidianDistance(m,n):

  # Array de indices de i
  i = np.arange(m)

  # Array de indices de j
  j = np.arange(n)

  # Matrix indexada
  x,y = np.meshgrid(i,j,indexing='ij')

  # Retorno do calculo da distancia euclidiana em relação a posiçao 0,0
  return np.sqrt((x-i[0])**2 + (y-j[0])**2)

print(euclidianDistance(5,4))
```

## Parte 5 - Geração automática de imagens sintéticas e manipulação de arquivos (opicional para EA979)

### Exercício 5.1

**a)** Quando geramos imagens sintéticas, estamos interessados em criar um conjunto de imagens com algumas características pré-definidas, mas com certa variabilidade. Para exemplificar esse processo, vamos automatizar a criação dos quadrados do exercício 3.1.

Crie uma função que tenha como entradas:
- Inteiro **n** que corresponde ao número de imagens a serem criadas;
- Inteiro **h** que corresponde à "altura" de cada imagem (com "largura" 2\***h**);

e que retorne:
- Um array **I** de shape (n, h, 2\*h, 3), em que cada **I**[n] corresponda a uma imagem similar à imagem gerada no exercício 3.1.b), mas colorida.

A ideia é que a função crie um conjunto de **n** imagens, todas com os "quadrados menores" aproximadamente centralizados nos "quadrados maiores". O tamanho dos "quadrados menores" deve ser igual para todas as imagens do conjunto e definido de forma aleatória. As duas cores utilizadas devem variar para cada imagem e também são definidas aleatoriamente.

Você deve definir certos limites para garantir que a geração das imagens faça sentido: qual a altura mínima das imagens? qual o tamanho mínimo dos quadrados menores? qual a distância mínima entre as bordas dos quadrados menor/maior (para que um não se sobreponha ao outro)?

Teste a função com diferentes valores de entrada, tentando sempre plotar as imagens de cada conjunto gerado em um único plot. Comente sobre as decisões tomadas na criação da função (como os limites que foram definidos) e seu comportamento esperado (como a centralização dos quadrados menores em relação às dimensões da imagem).

```python
def createImage(n,h):
  # Chega se o tamanho é menor que 6
  if h < 6:

    print("Insira valores maiores que 6")

  else:

    # Gera o tamanho do quadrado menor de forma aleatória com limite inferior como sendo um décimo de h e no máximo 80% de h
    x = np.random.randint(low=h/4, high = 0.8*h)//2

    # Criando o frame com 0s
    image= np.zeros([n, h, 2*h, 3], dtype=np.uint8)

    # Um loop for para pintar cada imagem
    for i in range(n):

      # Definindo array com cores aleatórias
      sqr_color = np.random.randint(0,256, size=(2, 3))

      # Pintando o primeiro quadrado
      image[i, :, :h] = sqr_color[0]

      # Pintando o segundo quadrado
      image[i, :, h:2*h] = sqr_color[1]

      # Pintando o primeiro quadradinho
      image[i, h//2 - x : h//2 + x, h//2 - x: h//2 + x] = sqr_color[1]

      # Pintando o segundo quadradinho
      image[i, h//2 - x : h//2 + x, 3*h//2 - x: 3*h//2 + x] = sqr_color[0]


    return image

# Criando as imagens a partir da função
images = createImage(5,12)

# Definindo o frame da imagem
fig, axs = plt.subplots(5, 1, figsize=(5,10))
fig.suptitle("Imagens de quadrados com quadradinhos")

# Plotando cada imagem em um subplot diferente
for i,ax in enumerate(axs.flat):

    plt.subplot(5, 1, 1+i)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(f"Image {i}")
plt.show()
```

```python
import random

# Criando vetor de valores aleatórios para a altura
height = np.random.randint(6,100, size = 4)

# Criando diferentes imagens
A = createImage(10,height[0])
B = createImage(10,height[1])
C = createImage(10,height[2])
D = createImage(10,height[3])

# Definindo nome do arquivo de saída
outfile = "squares.npz"
np.savez(outfile, A=A, B=B, C=C, D=D)

# Carregando imagem
npzfile = np.load(outfile)

# Definindo o frame da imagem
fig, axs = plt.subplots(1, 4, figsize=(10,2))
fig.suptitle("Imagens de quadrados com quadradinhos selecionados aleatoriamente")

# Plotando cada imagem em um subplot diferente utilizando enumerate para definiri o indice e a imagem correspondente
# Utilizando random.choice para selecionar aleatoriamente uma imagem de cada conjunto.
for i,x in enumerate(npzfile.files):
    plt.subplot(1, 4, 1+i)
    plt.imshow(random.choice(npzfile[x]))
    plt.axis('off')
    plt.title(f"Image {x}")
plt.show()
```

**b)** Utilize os dados extraídos no item anterior para localizar as teclas da calculadora. Plote linhas vermelhas horizontais e verticais sobre a imagem original, passando aproximadamente no meio das teclas. Comente os métodos utilizados.

```python
copia = imagem2.copy()

# Filtrando a cópia para valores muito altos, trocando elas por 1
# Filtrando as colunas
filterCol = np.where( copia.mean(axis=0) > 135, 0, 1)

# Filtrando as linhas
filterRow = np.where( copia.min(axis=1) < 30, 0, 1)

# Filtrando as colunas
filterCol2 = np.where( copia.mean(axis=0) > 135, 1, 0)

# Filtrando as linhas
filterRow2 = np.where( copia.min(axis=1) < 30, 1, 0)

# Recuperando os indices onde está o centro das teclas
idxCol = np.where( filterCol == 1)[0]
idxRow = np.where( filterRow == 1)[0]
idxCol = np.delete(idxCol, np.argwhere(np.ediff1d(idxCol) <= 10) + 1)
idxRow = np.delete(idxRow, np.argwhere(np.ediff1d(idxRow) <= 10) + 1)

# Recuperando os indices onde está o centro das teclas
idxCol2 = np.where( filterCol2 == 1)[0]
idxRow2 = np.where( filterRow2 == 1)[0]
idxCol2 = np.delete(idxCol2, np.argwhere(np.ediff1d(idxCol2) <= 10) + 1)
idxRow2 = np.delete(idxRow2, np.argwhere(np.ediff1d(idxRow2) <= 10) + 1)

# Calculando o centro de cada tecla
vert = (idxCol[idxCol!=0] + idxCol2[idxCol2!=0])//2
hor = (idxRow[idxRow!=1] + idxRow2[idxRow2!=1])//2

plt.figure(figsize=(2, 2))
# Adicionar linhas vermelhas horizontais nas posições médias das teclas
for indice in vert:
    plt.axvline(x=indice, color='red', linestyle='-', linewidth=0.5)

# Adicionar linhas vermelhas verticais nas posições máximas das teclas
for indice in hor:
    plt.axhline(y=indice, color='red', linestyle='-', linewidth=0.5)


plt.imshow(imagem2)
plt.axis('off')
plt.show()
```

**c)** Agora, plote sobre a imagem original um *bounding box* ao redor de cada tecla localizada. Cada *bounding box* deve ser um retângulo vermelho definindo as bordas da tecla. Comente os métodos utilizados.
